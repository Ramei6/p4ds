# Commment l'offre de transport évolue-t-elle avec la densité à Paris ?


## Sujet et plan

### Choix du sujet

**Pourquoi s'intéresser à la densité en zone urbaine ? L'optimisation des coûts** : De nos jours, les sujets de l'urbanisme et de l'organisation l'espace public correspondent à un enjeu d'efficacité de l'usage des ressources et d'efficacité économique. Le but est d'avoir une ville efficace (accès aux transports, espaces verts, bien-être...) tout en minimisant les coûts, empreintes écologiques inclues. Dans ce cadre, la densité de l'activité humaine urbaine est un facteur clef. 

Outre les enjeux d'artificialisation des sols que l'on connaît actuellement, le document [Les liens entre densité et transports publics : intérêt et limites de la notion de densité pour expliquer les pratiques de mobilité et le choix d’un système de TCSP, d'Aline Gillette (2014)](https://dumas.ccsd.cnrs.fr/dumas-02914337/document) dresse plusieurs constats. Plus la densité de population est importante, plus la part de ménages non-motorisés est importante et plus la part de ménages multi-motorisés est faible. (0.5 voitures par ménages à Paris contre plus de 1 dans la plupart des villes étudiées). Aussi, on peut lire page 39 que lorsque la densité de population passe de 10 000 hab/km2 à 30 000 hab/km2, environ, la part modale de la voiture est divisée par 3, passant de 30% à 10%, quand celle des transports en commun et de la marche croissent de 10% respectivement.

Or le coût que représente la possession et l'usage d'une voiture est important : dans l'indice des prix à la consommation de l'INSEE, le simple achat d'automobiles représente 3.95% du budget des ménages et leur usage 9.61% (Source : [Indice français des prix à la consommation](https://fr.wikipedia.org/wiki/Indice_fran%C3%A7ais_des_prix_%C3%A0_la_consommation)) (Ce dernier chiffre inclut l'usage des motocyclettes, mais l'ordre de grandeur demeure.). Ainsi, l'automobile peut consitituer de l'ordre de 15% du budget d'un ménage motorisé. Ce coût est d'autant plus élevé que le chiffre précédemment cité n'inclut pas les externalités négatives en termes de sédentarité et de pollution qui affectent les finances de la sécurité sociale ainsi que la capacité de production à long terme. Le modèle de Nordaus (Chapitre 3 de Macroéconomie 1) évalue par exemple dans certains cas une taxe carbone optimale (non appliquée actuellement) de plus de 500 dollars la tonne de CO2 en 2025. Le coût estimé pour la société du litre d'essence est dans ce cas de 500*2.28/1000 = 1.14$ et il n'est pas pris en compte dans les taxes actuelles, lesquelles ne compensent d'ailleurs pas le coût de l'usure des routes pour la collectivité (Source : [bilan annuel des transports en 2023](https://www.bnsp.insee.fr/ark:/12148/bc6p09kb7p8.pdf)).

Ce faisant, l'enjeu de la très haute densité et de son aménagement en termes de transports notamment est majeur sur le plan économique. Nous allons donc nous questionner sur les liens qui peuvent être faits entre densité et transports en zone très dense.

**Choix de Paris** : D'une part, Paris se caractérise par une grande densité d'activité sur une surface importante. D'autre part, une grande quantité de données à son sujet est disponible, notamment [celles du site du département](https://opendata.paris.fr/pages/home/). Voici donc la raison pour laquelle Paris a été choisie pour une étude des hautes densités. On présume que la densité de l'activité humaine urbaine va être corrélée à l'offre de transport, avec a priori en zone plus dense une moindre surface de chaussée par unité d'activité humaine, une plus forte densité en transport en commun, une infrastructure donnant plus de place aux deux roues, lesquels sont moins consommateurs d'espace. Cependant, les [politiques publiques jouent aussi dans la part allouée aux différents modes de transports](https://cst.episciences.org/12161/pdf), ce qui devrait nous permettre d'observer d'autres différences dans l'offre de transport au-delà de la densité.

**Problématique** : Ainsi, on va se demander commment l'offre de transport évolue-t-elle avec la densité à Paris ?
