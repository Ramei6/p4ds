# Commment l'offre de transport évolue-t-elle avec la densité à Paris ?


## Sujet et plan

### Choix du sujet

**Pourquoi s'intéresser à la densité en zone urbaine ? L'optimisation des coûts** : De nos jours, les sujets de l'urbanisme et de l'organisation l'espace public correspondent à un enjeu d'efficacité de l'usage des ressources et d'efficacité économique. Le but est d'avoir une ville efficace (accès aux transports, espaces verts, bien-être...) tout en minimisant les coûts, empreintes écologiques inclues. Dans ce cadre, la densité de l'activité humaine urbaine est un facteur clef. 

Outre les enjeux d'artificialisation des sols que l'on connaît actuellement, le document [Les liens entre densité et transports publics : intérêt et limites de la notion de densité pour expliquer les pratiques de mobilité et le choix d’un système de TCSP, d'Aline Gillette (2014)](https://dumas.ccsd.cnrs.fr/dumas-02914337/document) dresse plusieurs constats. Plus la densité de population est importante, plus la part de ménages non-motorisés est importante et plus la part de ménages multi-motorisés est faible. (0.5 voitures par ménages à Paris contre plus de 1 dans la plupart des villes étudiées). Aussi, on peut lire page 39 que lorsque la densité de population passe de 10 000 hab/km2 à 30 000 hab/km2, environ, la part modale de la voiture est divisée par 3, passant de 30% à 10%, quand celle des transports en commun et de la marche croissent de 10% respectivement.

Or le coût que représente la possession et l'usage d'une voiture est important : dans l'indice des prix à la consommation de l'INSEE, le simple achat d'automobiles représente 3.95% du budget des ménages et leur usage 9.61% (Source : [Indice français des prix à la consommation](https://fr.wikipedia.org/wiki/Indice_fran%C3%A7ais_des_prix_%C3%A0_la_consommation)) (Ce dernier chiffre inclut l'usage des motocyclettes, mais l'ordre de grandeur demeure.). Ainsi, l'automobile peut consitituer de l'ordre de 15% du budget d'un ménage motorisé. Ce coût est d'autant plus élevé que le chiffre précédemment cité n'inclut pas les externalités négatives en termes de sédentarité et de pollution qui affectent les finances de la sécurité sociale ainsi que la capacité de production à long terme. Le modèle de Nordaus (Chapitre 3 de Macroéconomie 1) évalue par exemple dans certains cas une taxe carbone optimale (non appliquée actuellement) de plus de 500 dollars la tonne de CO2 en 2025. Le coût estimé pour la société du litre d'essence est dans ce cas de 500*2.28/1000 = 1.14$ et il n'est pas pris en compte dans les taxes actuelles, lesquelles ne compensent d'ailleurs pas le coût de l'usure des routes pour la collectivité (Source : [bilan annuel des transports en 2023](https://www.bnsp.insee.fr/ark:/12148/bc6p09kb7p8.pdf)).

Ce faisant, l'enjeu de la très haute densité et de son aménagement en termes de transports notamment est majeur sur le plan économique. Nous allons donc nous questionner sur les liens qui peuvent être faits entre densité et transports en zone très dense.

**Choix de Paris** : D'une part, Paris se caractérise par une grande densité d'activité sur une surface importante. D'autre part, une grande quantité de données à son sujet est disponible, notamment [celles du site du département](https://opendata.paris.fr/pages/home/). Voici donc la raison pour laquelle Paris a été choisie pour une étude des hautes densités. On présume que la densité de l'activité humaine urbaine va être corrélée à l'offre de transport, avec a priori en zone plus dense une moindre surface de chaussée par unité d'activité humaine, une plus forte densité en transport en commun, une infrastructure donnant plus de place aux deux roues, lesquels sont moins consommateurs d'espace. Cependant, les [politiques publiques jouent aussi dans la part allouée aux différents modes de transports](https://cst.episciences.org/12161/pdf), ce qui devrait nous permettre d'observer d'autres différences dans l'offre de transport au-delà de la densité.

**Problématique** : Ainsi, on va se demander commment l'offre de transport évolue-t-elle avec la densité à Paris ?

**Démarche vis-à-vis des données initiales** : les données que nous utilisons sont de très bonne qualité, mais sont brutes, ce qui nécessite un travail de formatage et de réflexion sur celui-ci : quelle échelle choisir, quelle surface considérer, quelle variable doit être produite à des fins d'analyse, que signifie-t-elle exactement ? Par exemple, avant d'enlever les espaces verts, voies ferrées et voies d'eau, certaines densités de bâti sont excessivement parce que l'on divise principalement par l'aire d'une surface non-bâtie en cas de présence d'un bois dans l'aire géographique.

### Plan
**I** : Tout d'abord, il faut s'intéresser à la mesure de la densité et plus précisément à la question de la surface à laquelle rapporter telle ou telle données. Nous allons utiliser plusieurs types de zones géographiques (arrondissements, quartiers) plus ou moins précises et nous intéresser à la correction de celles-ci, notamment par l'exclusion des zones qui ne peuvent pas être bâties et ne peuvent ce faisant pas accueillir de forte activité humaine urbaine, ce qui peut fausser les calculs de densité. La fin de [ce document](https://www.persee.fr/doc/canan_0767-8436_2002_num_58_1_1068) donne un exemple de la nécessité de ces corrections et de ces réflexions.

**II** : Ensuite, il faut s'intéresser à l'indicateur d'activité humaine urbaine. On peut penser à la densité de population, la densité d'emploi, leur somme, ou bien à la densité de bâti, définie comme le ratio de la superficie de bâtiment utilisable sur la superficie au sol. La réflexipon sur ce thème menée, sera choisi un indicateur de la densité d'activité humaine pour la suite de l'étude.

**III** : Après, il faut s'intéresser à l'offre de transport en elle-même.

**III.A** : Dans un premier temps, on étudie les moyens de transport individuels.

On étudie d'abord la surface de chaussée nécessaire à la desserte d'une unité d'activité humaine, par exemple 1 000 m2 de bâti. On mesure la superficie de chaussée pour chaque zone géographique en prenant garde à retrancher les surfaces de stationnement. Puis on la rapporte à l'indicateur d'activité humaine urbaine. Après avoir obtenu la cartographie du ratio de la superficie de chaussée par unité d'activité urbaine, on étudie le lien entre celui-ci et la densité d'activité humaine. On s'attend à une relation négative étant donné que la part modale de la voiture diminue avec la densité, y compris entre 10 000 hab/km2 et 30 000 hab/km2 (cf paragraphe précédent.). On mène cette étude a priori à toutes les échelles afin de saisir de potentielles disparités.
**ACCESSOIRE On étudie aussi à l'échelle de l'arrondissement et du quartier, une échelle plus fine posant des problèmes de mesure, la densité de maillage routier, définie en nombre de kilomètres de voiries par kilomètre carré au sol. Cette variable donne une indication sur la structure de la desserte vicinale et le type d’urbanisme. En effet, on peut se demander si pour une même densité de bâti, il est préférable d'avoir une multitude de petites rues et de petits îlots urbains (=pâtés de maison), c'est-à-dire une grande densité de maillage routier, afin de minimiser la densité de chaussée, ou s'il est préférable d'avoir quelques larges rues et de grands îlots urbains. On étudie alors le lien entre densité de maillage routier et ratio superficie de chaussée/activité humaine.**

On étudie ensuite la qualité de l'infrastructure pour les 2 roues, a priori plus économe en espaces et donc plus adapté aux grands espaces. On étudie donc le nombre de stationnement 2 roues par unité de volume bâti et étudions le lien avec la densité de bâti. On étudie aussi la part de la surface de stationnement en extérieur allouée aux 2 roues et étudions le lien avec la densité de bâti. On réalise ces études à toutes les échelles (arrondissements, quartier, iris). On étudie aussi le ratio du nombre de kilomètre de voies cyclables sur le nombre de kilomètres total de voies pour mesurer l'aspect cyclables ou non des différentes zones géographiques et son lien avec la densité, cette fois-ci seulement à l'échelle de l'arrondissement et du quartier, la prise en compte des bords faussant les données pour les zones plus petites. Nous présumons que les politiques publiques étant adaptées, les relations seront croissantes. Ce n'est pas forcément le cas.

## II Mesure de la densité d'activité humaine

A priori, on peut s'intéresser à plusieurs densité afin de mesurer la densité de l'activité humaine urbaine. La densité de population, la densité d'emploi, leur somme et la densité du bâti, définie comme la surface de bâtiments utilisable par unité de surface au sol.

D'un côté, concernant les données de population, on peut s'appuyer sur la base de données de l'INSEE datant de 2022 [Population en 2022 Recensement de la population - Base infracommunale (Iris)](https://www.insee.fr/fr/statistiques/8647014#dictionnaire) pour obtenir une première approximation de l'activité humaine urbaine. Avant de faire l'effort de traiter les données, nous avons souhaité vérifer sur des cartes déjà publiées si la densité de population pouvait être une variable pertinente pour mesurer l'activité. D'un autre côté, concernant les données d'emploi, elles sont disponibles sur la base [SIRENE de l'INSEE](https://www.data.gouv.fr/dataservices/api-sirene-open-data). Ce faisant, on peut se rabattre sur [cette carte publiée en 2021 sur le site de l'APUR (atelier parisien d'urbanisme)](https://www.apur.org/sites/default/files/documents/cartefichiers-attaches/carte_densite_humaine_population_emplois_salaries_mgp.pdf?openfile_analytics=770) qui en sommant habitants et emplois par kilomètre carré au niveau des IRIS nous montre notamment une très forte activité dans le quart nord-ouest du centre-ville, cela invalide immédiatement l'usage de la densité de population. Malheureusement, l'accès à cette base nécessite l'existence d'un compte dont nous ne disposons a priori pas.

L’hypothèse principale est celle d’une relation croissante : les zones de forte densité d’activité humaine devraient bénéficier d’une densité plus élevée d’équipements de transport collectif, cohérente avec la part modale plus importante de ces modes dans les zones denses.

Dans un second temps, on adopte un point de vue complémentaire en rapportant le nombre d’équipements de transport collectif non plus à la surface au sol, mais à la surface de bâti utilisable. Leur relation avec la densité du bâti est a priori ambiguë : une forte densité peut permettre de desservir davantage d’activité avec un nombre relativement limité d’équipements, traduisant une meilleure efficacité spatiale du système de transport.

#### Mobilités partagées : taxis et vélos en libre-service

Enfin, on s’intéresse aux formes intermédiaires de mobilité que constituent les stations de taxis et les stations de vélos en libre-service. Ces modes, situés à la frontière entre transport individuel et transport collectif, sont susceptibles de jouer un rôle important dans les zones de forte densité, où la demande est concentrée et les distances moyennes plus courtes.

À partir des bases de données relatives aux bornes d’appel taxi et aux stations Vélib’, on calcule leur densité spatiale et leur densité rapportée à la surface de bâti utilisable. On étudie ensuite leur relation avec la densité de l’activité humaine urbaine afin d’évaluer si ces dispositifs sont prioritairement déployés dans les zones les plus denses.
