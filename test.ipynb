```python
import pandas as pd

def InArrondissement(nb_arrondissement):
    df_arrets = pd.read_csv(
        path + "arrets.csv",
        sep=';',
        usecols=[
            'ArRName', 'ArRType', 'ArRTown',
            'ArRPostalRegion', 'ArRAccessibility',
            'ArRFareZone', 'ArRGeopoint'
        ]
    )

    # Sécurisation + extraction de l’arrondissement
    mask = (
        df_arrets["ArRPostalRegion"]
        .astype(str)
        .str.match(r"^75") &
        (df_arrets["ArRPostalRegion"].astype(str).str[-2:].astype(int) == nb_arrondissement)
    )

    return df_arrets.loc[mask]

InArrondissement(1)
```
