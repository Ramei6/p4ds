## III Ce que nous aurions aimé faire : notre projet initial d'étude du lien entre la densité d'activité humaine et l'offre de transports

Nous nous sommes restreints à l'étude de la densité du bâti, mais nous souhaitions étudier son lien avec l'offre de transports.

De nos jours, les sujets de l'urbanisme et de l'organisation l'espace public correspondent à un enjeu d'efficacité de l'usage des ressources et d'efficacité économique. Le but est d'avoir une ville efficace (accès aux transports, espaces verts, bien-être...) tout en minimisant les coûts, empreintes écologiques inclues. Dans ce cadre, la densité de l'activité humaine urbaine est un facteur clef. 

Outre les enjeux d'artificialisation des sols que l'on connaît actuellement, le document [Les liens entre densité et transports publics : intérêt et limites de la notion de densité pour expliquer les pratiques de mobilité et le choix d’un système de TCSP, d'Aline Gillette (2014)](https://dumas.ccsd.cnrs.fr/dumas-02914337/document) dresse plusieurs constats. Plus la densité de population est importante, plus la part de ménages non-motorisés est importante et plus la part de ménages multi-motorisés est faible. (0.5 voitures par ménages à Paris contre plus de 1 dans la plupart des villes étudiées). Aussi, on peut lire page 39 que lorsque la densité de population passe de 10 000 hab/km2 à 30 000 hab/km2, environ, la part modale de la voiture est divisée par 3, passant de 30% à 10%, quand celle des transports en commun et de la marche croissent de 10% respectivement.

Or le coût que représente la possession et l'usage d'une voiture est important : dans l'indice des prix à la consommation de l'INSEE, le simple achat d'automobiles représente 3.95% du budget des ménages et leur usage 9.61% (Source : [Indice français des prix à la consommation](https://fr.wikipedia.org/wiki/Indice_fran%C3%A7ais_des_prix_%C3%A0_la_consommation)) (Ce dernier chiffre inclut l'usage des motocyclettes, mais l'ordre de grandeur demeure.). Ainsi, l'automobile peut consitituer de l'ordre de 15% du budget d'un ménage motorisé. Ce coût est d'autant plus élevé que le chiffre précédemment cité n'inclut pas les externalités négatives en termes de sédentarité et de pollution qui affectent les finances de la sécurité sociale ainsi que la capacité de production à long terme. Le modèle de Nordaus (Chapitre 3 de Macroéconomie 1) évalue par exemple dans certains cas une taxe carbone optimale (non appliquée actuellement) de plus de 500 dollars la tonne de CO2 en 2025. Le coût estimé pour la société du litre d'essence est dans ce cas de 500*2.28/1000 = 1.14$ et il n'est pas pris en compte dans les taxes actuelles, lesquelles ne compensent d'ailleurs pas le coût de l'usure des routes pour la collectivité (Source : [bilan annuel des transports en 2023](https://www.bnsp.insee.fr/ark:/12148/bc6p09kb7p8.pdf)).

Ce faisant, l'enjeu de la très haute densité et de son aménagement en termes de transports notamment est majeur sur le plan économique. Nous allons donc nous questionner sur les liens qui peuvent être faits entre densité et transports en zone très dense.

**Choix de Paris** : D'une part, Paris se caractérise par une grande densité d'activité sur une surface importante. D'autre part, une grande quantité de données à son sujet est disponible, notamment [celles du site du département](https://opendata.paris.fr/pages/home/). Voici donc la raison pour laquelle Paris a été choisie pour une étude des hautes densités. On présume que la densité de l'activité humaine urbaine va être corrélée à l'offre de transport, avec a priori en zone plus dense une moindre surface de chaussée par unité d'activité humaine, une plus forte densité en transport en commun, une infrastructure donnant plus de place aux deux roues, lesquels sont moins consommateurs d'espace. Cependant, les [politiques publiques jouent aussi dans la part allouée aux différents modes de transports](https://cst.episciences.org/12161/pdf), ce qui devrait nous permettre d'observer d'autres différences dans l'offre de transport au-delà de la densité.

**Problématique initiale du projet** : On souhait se demander quel était le lien entre l'offre de transport et la densité à Paris ?

Notre point important est le coût du transport. En ce qui concerne l'activité humaine urbaine, on veut vouloir garantir une même qualité de desserte pour un coût minimal. Dans ce cadre, la densité de l'activité est un facteur clé. Une haute densité réduit les surfaces à desservir et les distances.

### III.A Notre projet d'étude concernant la desserte en transport en commun

Nous partions du principe que la densité était bénéfique pour la qualité de la desserte en transports en commun. Si l'on souhaite par exemple disposer une station de transport en commun pour chaque carré de 500m par 500m, au plus l'activité est dense, c'est-à-dire se répartit sur une petite surface, au moins il faut bâtir et entretenir un nombre important de stations ainsi qu'un réseau routier ou ferroviaire important, au plus le coût de desserte de l'activité totale est réduit a priori. De plus, si l'on souhaite plutôt disposer une station de transport en commun pour chaque unité d'activité humaine, par exemple de 10 000 habitants, au plus cette activité est concentrée, dense, au plus cette activité est proche de la station de transport en commun, ce qui améliore la qualité de la desserte. Par ailleurs, a priori, au plus une activité importante se déroule à proximité d'une station de transport en commun, au plus elle va être utilisée, ce qui permet d'augmenter la fréquence de la desserte de la station et donc d'améliorer la qualité de la desserte. Cette [étude](https://dumas.ccsd.cnrs.fr/dumas-02914337/document) indiquait bien un lien fort et une corrélation positive entre part modale des transports en commun et densité. (La part modale étant la porportin de personnes utilisant ce mode de transport.)
De là, nous prévoyions d'étudier le lien entre densité de bâti et qualité de la desserte. Nous nous demandions également si le clivage Est/Ouest politique apparaîtrait, avec davantage de transports en commun à l'Est parisien, plus orienté à gauche.

#### III.A.1 Etude prévue du réseau de métropolitain

Pour cela, en plus de disposer de nos zones géographiques (arrondissements, quartiers) corrigées des espaces verts, des voies ferrées et des voies d'eau, des volumes bâtis par zone géographique, et donc des **densités de bâtis corrigées**, nous comptions disposer du nombre de stations de bus par zone géographique corrigée.

Notre premier objectif était d'obtenir la **densité en station de métro**, définie comme le nombre de station de métro par kilomètre carré de surface n'étant ni un espace vert, ni une voie d'eau, ni une voie ferrée.
Nous souhaitions étudier cette variable car plus elle est élevée, plus l'activité est proche des stations de métro, mieux elle est déservie.
De là, nous souhaitions obtenir deux premières cartes interactives concernant respectivement les arrondissements et les quartiers (l'étude précédente nous montrant que les IRIS constituait une échelle trop fine) affichant la **densité en station de métro**.
Puis nous souhaitions tracer le nuage de points avec en abscisse la **densité de bâti** et en ordonnée la **densité en station de métro** avant de réaliser une régression.
Nous nous attendions, comme indiqué dans cette [étude](https://dumas.ccsd.cnrs.fr/dumas-02914337/document), à obtenir une relation positive entre ces deux densités, ce qui signifie que plus la **densité de bâti** augmente, plus l'activité se trouve proche d'une station, donc meilleure est la desserte de l'activité.
On s'attend à cette relation positive car a priori, plus une zone est dense, plus les pouvoirs sont incités à bien le desservir en transports en commun car ceux-ci seront plus souvent utilisés.

Notre second objectif était d'obtenir le **ratio de desserte en station de métro**, défini comme le nombre de station de bus par mètre carré de plancher/de bâti.
Plus il est bas, plus une station dessert d'activité, donc plus le coût de desserte d'une unité d'activité diminue.
De là, nous souhaitions obtenir deux autres cartes interactives concernant respectivement les arrondissements et les quartiers (l'étude précédente nous montrant que les IRIS constituait une échelle trop fine) affichant le **ratio de desserte en station de métro**, notamment pour observer un attendu clivage est-ouest, avec un est mieux doté a priori que l'ouest.
Puis nous souhaitions tracer le nuage de points avec en abscisse la **densité de bâti** et en ordonnée le **ratio de desserte en station de métro** avant de réaliser une régression.
Nous espérions obtenir un résultat intéressant.
D'un côté, l'on peut penser que si les pouvoirs publics maintiennent une **densité en station de métro** par convenance, on doit alors s'attendre à une diminution de ce **ratio de desserte en station de métro** avec l'augmentation de la **densité de bâti**.
D'un autre côté, une plus forte **densité de bâti** peut entraîner un effet boule de neige, entraînant les pouvoirs publics à metre davantage de transports en commun à disposition, entraînant la population à davantage le prendre, etc.

Nous comptions pour cela utiliser la base de données de la région Ile-de-France [Référentiel des arrêts : Arrêts :](https://data.iledefrance-mobilites.fr/explore/dataset/arrets/table/). Cette base de données permet de géolocaliser chaque arrêt de bus, de métro, de tramway. On peut utiliser ArRName pour compter le nombre d'apparition d'une station, puis créer un poids permettant de pondérer la valeur de chaque arrêt de façon à ce qu'une station ait un poids aggrégé de 1 et que le fait qu'une sortie de station soit d'un côté ou de l'autre d'une rue ne fausse pas les densités, en l'attribuant à une zone géographique plutôt qu'une autre.

### III.B Notre projet d'étude concernant la desserte en transport à usage individuel mais partagé

A la frontière des transports en commun, on peut placer les réseaux de transports à usage individuel comme le taxi ou les stations de vélos en libre-service. On peut présumer qu'au plus la densité de bâti est importante, au plus ces solutions sont proposées car elles sont plus économes en place. Par exemple, si [ce document](https://droitauvelo.org/IMG/pdf/heurovelo54.pdf?utm_source=chatgpt.com) peut paraître assez daté et engagé, il cite le commissariat au plan pour ses chiffres de sa page 9, lesquels sont un débit maximal de 1 000 personnes par heure pour une piste cyclable unidirectionnelle monovoie, soit 1000 personnes/heure/mètre de largeur pour un coût au kilomètre pour un débit de 1000 personnes heures de 0.02 millions d'euros, contre un débit maximal de 2000 personnes par heure pour une 4 voies de contournement sans carrefours, soit 2000/(4*3) = 166 personnes/heure/mètre de largeur de voie pour un coût au kilomètre pour un débit de 1000 personnes heures de 10/2 = 5 millions d'euros, soit 250 fois plus. De plus, un taxi est toujours en mouvement, quand une voiture individuelle occupe une surface de stationnement l'ultra-mmajorité du temps.

Nous comptions appliquer exactement la même méthode que précédemment avec les stations vélibs et les stations de taxis, avions les mêmes présomptions pour le premier tracé et les mêmes interrogations pour le second. Cependant, le simple coup d'oeil sur les bases de données [Vélib' - Localisation et caractéristique des stations](https://opendata.paris.fr/explore/dataset/velib-emplacement-des-stations/information/) et [Stations de taxi ](https://opendata.paris.fr/explore/dataset/bornes-dappel-taxi/information/) nous faisait remarquer qu'il y avait une surdotation en taxi à l'ouest, que nous présumerions due à une plus forte densité d'emploi et un niveau de vie plus élevé, et une surdotation en Vélib dans le nord du centre-ville. Aussi, nous étions curieux du résultat.

### III.C Notre projet d'étude concernant la desserte en transport individuel

Nous souhaitions observer comment évoluait l'offre de transport individuels avec la densité et notamment si les pouvoirs publics favorisaient en zone dense les modes de transport économes en espace comme les cycles.

#### III.C.1 Zone piétonne

On ne souhaitait pas s'intéresser à la présence de zones piétonnes car le développement des super-blocks à Paris rend progressivement flou le caractère piétonnier ou non d’une rue, les rues à l’intérieur d’un [super-bloc] (https://www.youtube.com/watch?v=ZORzsubQA_M) étant de facto quasiment réservées aux piétons et aux vélos.

#### III.C.2 Etude prévue des surfaces de chaussées

Relativement à l'entretien des chaussées, a priori, au plus une activité est dense, au plus le nombre de mètres carré de chaussée nécessaires pour le desservir est faible, ce qui peut permettre d'en diminuer le coût d'entretien par unité d'activité humaine.
Nous souhaitions donc aggréger l'ensemble des surfaces de chaussées issue de [Plan de Voirie – chaussée :](https://opendata.paris.fr/explore/dataset/plan-de-voirie-chaussees/map/?disjunctive.num_pave&basemap=jawg.dark&location=17,48.86156,2.40738), puis y retrancher les emprises de stationnement, issues de la base [Stationnement sur voie publique – emprises : ](https://opendata.paris.fr/explore/dataset/stationnement-sur-voie-publique-emprises/map/?disjunctive.regpri&disjunctive.regpar&disjunctive.typsta&disjunctive.arrond&disjunctive.locsta&disjunctive.zoneres&disjunctive.parite&disjunctive.signhor&disjunctive.signvert&disjunctive.confsign&disjunctive.typemob&basemap=jawg.dark&location=16,48.85864,2.38511), moins coûteuses à entretenir, puis en faire l'intersection avec les zones géographiques excluant espaces verts, voies ferrées et voies d'eau pour obtenir le **ratio de desserte en chaussée** en mètre carré de chaussée par mètre carré de plancher. Plus il est élevé, plus l'entretien du réseau routier est cher. 
Plus la densité augmente, moins les distances à parcourir sont grandes, donc moins il y a besoin de voirie, donc on présumait une diminution du **ratio de desserte en chaussée** avec la hausse de la **densité du bâti**.
Ainsi, nous souhaitions tracer le nuage de points et réaliser la régression pour vérifier cette hypothèse.
Nous souhaitions également obtenir la carte interactive des **ratio de desserte en chaussée** pour les arrrondissements et les quartiers (desquels on a retranché les esapces verts et autres) afin d'obtenir une vision d'ensemble de cette variable et peut-être obtenir un clivage entre le centre historique et les parties plus récentes de la ville.

#### III.C.3 Etude prévue de la densité du maillage routier 

Cependant, il nous semblait important d'introduire une troisième notion de **densité de maillage routier**, définie en nombre de kilomètres de voiries par kilomètre carré au sol.
Nous comptions utiliser la base [Tronçons de voie](https://opendata.paris.fr/explore/dataset/troncon_voie/information/) et aggréger les linéaires de voies par zone géographique (arrondissements et quartiers) corrigée des espaces verts et autres.
Cette variable donne une indication sur la structure de la desserte vicinale et le type d’urbanisme. En effet, on peut se demander si pour une même densité de bâti, il est préférable d'avoir une multitude de petites rues et de petits îlots urbains (=pâtés de maison), c'est-à-dire une grande densité de maillage routier, ou s'il est préférable d'avoir quelques larges rues et de grands îlots urbains, afin de minimiser le **ratio de desserte en chaussée**. 
L'objectif était donc de trouver la structure de desserte et de bâti permettant de minimiser le **ratio de desserte en chaussée**.
Pour cela, nous comptions tracer le nuage de point avec en ordonnée le **ratio de desserte en chaussée** et en abscisse **densité de maillage routier** puis réaliser une régression.
Nous souhaitions également obtenir la carte interactive aux échelles des arrondissements et des quartiers de la **densité de maillage routier** afin d'avoir une idée de sa répartition et des typologies urbaines de Paris.
 

#### III.C.4 Etude prévue de la densité de piste cyclable

Nous comptions également utiliser [Linéaires d'aménagements cyclables :](https://opendata.paris.fr/explore/dataset/amenagements-cyclables/map/?disjunctive.arrondissement&disjunctive.position_amenagement&disjunctive.vitesse_maximale_autorisee&disjunctive.source&disjunctive.amenagement&location=16,48.85687,2.35247&basemap=jawg.streets) afin d'aggréger les linéaires d'aménagements cyclables par zone géographique (arrondissements et quartiers) corrigée des espaces verts et autres.
De là, nous souhaitions obtenir un "**ratio d'accessibilité cyclable**", défini par le rapport du nombre de kilomètres de pistes cyclables sur le nombre de kilomètres de voies (obtenu à l'aide de [Tronçons de voie](https://opendata.paris.fr/explore/dataset/troncon_voie/information/)) pour une zone géographique donnée.
Ce ratio est intéressant car plus celui-ci est élevé, plus a priori un itinéraire peut-être réalisé à vélo à l'abri de la circulation motorisée, donc plus l'aménagement encourage la pratique du vélo, plus économe en espace.
D'une part, nous souhaitions réaliser la carte interactive de ce ratio aux échelles des arrondissements et des quartiers pour observer un attendu clivage est-ouest avec un ouest moins doté et aussi obtenir une vision d'ensemble de cette variable à l'échlle de la ville.
D'autre part, nous souhaitions faire la régression de ce ratio sur la **densité de bâti** afin de savoir si les pouvoirs publics favorisaient davantage ce moyen de transport économe en espace dans les zones plus denses.

#### III.C.5 Etude prévue de l'offre de stationnement pour cycles

Finalement, on souhaitait savoir si les pouvoirs publics adaptaient ou non l'espace public en cas de fortes densités en donnant une plus grande part de la surface de stationnement aux moyens de transport individuels économes en place comme les motocyclettes et bicyclettes et en enfouissant davantage de parkings dédiés à l'automobile. Ceci nous amenait à vouloir considérer la base [Stationnement sur voie publique – emprises : ](https://opendata.paris.fr/explore/dataset/stationnement-sur-voie-publique-emprises/map/?disjunctive.regpri&disjunctive.regpar&disjunctive.typsta&disjunctive.arrond&disjunctive.locsta&disjunctive.zoneres&disjunctive.parite&disjunctive.signhor&disjunctive.signvert&disjunctive.confsign&disjunctive.typemob&basemap=jawg.dark&location=16,48.85864,2.38511), à sommer les surfaces de stationnements pour les deux roues d'une part et d'autre part sommer l'ensemble des surfaces de stationnement afin de faire le ratio de la première somme divisé par la seconde afin d'obtenir un **ratio de stationnement pour cycles**, et ceci à l'échelle des quartiers et des arrondissements (toujours en corrigeant les surfaces).
De là, on souhaitait obtenir les cartes interactives du **ratio de stationnement pour cycles** pour observer ou non un clivage est-ouest ou centre-périphérie.
On souhaitait aussi réaliser la régression du **ratio de stationnement pour cycles** sur la **densité de bâti**, ainsi que tracer le nuage de points pour les arrondissements et pour les quartiers, afin de là encore savoir si les pouvoirs publics favorisaient davantage ce moyen de transport économe en espace dans les zones plus denses.

### III.D Notre souhait d'utiliser une méthode clustering

Une fois obtenues les différentes variables décrivant l'offre de transport (**densité en station de métro**, **ratio de desserte en station de métro**, **densité en station de bus**, **ratio de desserte en station de bus**, **ratio de desserte en chaussée**, **densité de maillage routier**, **ratio d'accessibilité cyclable**, **ratio de stationnement pour cycles**) dans une base de données aggrégées à l'échelle des arrondissements, nous souhaitions réaliser une opération de clustering en cherchant de séparer l'ensemble des arrondissement en plusieurs clusters regroupant a priori des arrondissments à offre de transport similaire.
Nous souhaitions alors d'une part afficher ces différents clusters sur une carte afin de savoir l'on observait des clivages est-ouest ou centre-périphérie.  
Nous souhaitions d'autre part, en superposant les distributions de **densité de bâti** pour les différents clusters, vérifier si la densité de bâti permettait de caractériser ces clusters, c'est-à-dire l'offre de transport.

Nous souhaitions également réaliser cette étude à l'échelle des quartiers, plus nombreux, donc peut-être plus adaptés pour une opération de clustering.
